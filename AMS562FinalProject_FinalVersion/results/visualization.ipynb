```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
regular_customers_file = 'regular_customers.csv'
combined_customers_file = 'combined_customers.csv'

# Read the CSV files into pandas DataFrames
regular_data = pd.read_csv(regular_customers_file)
combined_data = pd.read_csv(combined_customers_file)

# Prepare data
regular_data = regular_data[regular_data['Day'] != 'Total']
combined_data = combined_data[combined_data['Day'] != 'Total']

regular_data['Day'] = regular_data['Day'].astype(int)
combined_data['Day'] = combined_data['Day'].astype(int)

# Exclude outliers
start_day = 10
end_day = 50

regular_data_filtered = regular_data[(regular_data['Day'] >= start_day) & (regular_data['Day'] <= end_day)]
combined_data_filtered = combined_data[(combined_data['Day'] >= start_day) & (combined_data['Day'] <= end_day)]

# Bar graph
plt.figure(figsize=(12, 6))
plt.bar(regular_data_filtered['Day'], regular_data_filtered['TotalDistance'], width=0.4, label='Regular Customers')
plt.bar(combined_data_filtered['Day'] + 0.4, combined_data_filtered['TotalDistance'], width=0.4, label='Combined Customers')
plt.title('Daily Total Distance Visualization (Excluding Outliers)')
plt.xlabel('Day')
plt.ylabel('Total Distance')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
